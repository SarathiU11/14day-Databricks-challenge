task -1 Convert raw CSV to Delta format.

```python
df = spark.read.format("csv")\
    .option("header", "true")\
    .option("inferSchema", "true")\
    .load("/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv")

```

```python
display(df.limit(5))
```

```python
df.write.format("delta") \
    .mode("overwrite") \
    .save("/Volumes/workspace/ecommerce/ecommerce_data/delta")
```

Task -2 Create a managed Delta table.

```python
df.write.format("delta").saveAsTable("event_table")
```

Task -3   Append data multiple times



```python
df_new = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("/Volumes/workspace/ecommerce/ecommerce_data/2019-Nov.csv")

df_new.write.format("delta") \
    .mode("append") \
    .saveAsTable("event_table")
```

```python
%sql

select count(*) from event_table
```

task - 4 Apply OPTIMIZE and observe improvement.

```python
%sql
OPTIMIZE event_table;
```
